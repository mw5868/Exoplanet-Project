# Using NASA Exoplanet Archive API to return Data

```python
from astropy.table import Table, Column
import matplotlib.pyplot as plt
url = "https://exoplanetarchive.ipac.caltech.edu/cgi-bin/nstedAPI/nph-nstedAPI?table=exoplanets&select=pl_hostname,ra,dec&order=dec&format=csv"
# This API returns Hostname, RA and Dec

t = Table.read(url, format="csv")

# Generate a plot of RA against Dec for the data.
plt.scatter(t["ra"],t["dec"])
plt.xlabel("RA")
plt.ylabel("DEC")
plt.title("Exoplanet Archive API Demo")
```

```python
url = "https://exoplanetarchive.ipac.caltech.edu/cgi-bin/nstedAPI/nph-nstedAPI?table=exoplanets"

t = Table.read(url,format="csv")
t_transit = t[t["pl_discmethod"]=="Transit"] # Stores all entries where discovery metohd is "Transit"

fig = plt.figure()
ax = fig.add_subplot(1,1,1,aspect="equal")
ax.scatter(t["ra"],t["dec"],color="Black")
ax.scatter(t_transit["ra"],t_transit["dec"],color="red")
ax.set_xlim(360,0)
ax.set_ylim(-90,90)
ax.set_ylabel("DEC")
ax.set_xlabel("RA")
ax.set_title("Positions of planets detected via Transit")
```
